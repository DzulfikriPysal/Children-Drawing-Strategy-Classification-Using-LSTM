```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Embedding
import sys
import csv
import tensorflow as tf
from tensorflow.contrib import learn
from matplotlib import pyplot
from keras.callbacks import History
from keras.layers import Dense, LSTM, Dropout
from keras.layers.embeddings import Embedding
import matplotlib.pyplot as plt
import time as time
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
tf.logging.set_verbosity(tf.logging.ERROR)
import numpy as np
```

```python
LabelEncoder = LabelEncoder()
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

print("Loading data")
training_input_data = csv.reader(open('z2.csv'), delimiter='-', quotechar='|')

print("done")
```

```python
from sklearn.model_selection import train_test_split

f=[]
label=[]
data=[]

for row in training_input_data:
	g = ",". join(str(elt) for elt in row)
	f.append(g)
	
	
for rows in f:
	g = rows
	g = g.split(",")
	a = g[0]
	del g[0]
	data.append(", ".join(str(elt) for elt in g))
	label.append(a)
	
x_train, x_test, y_train, y_test = train_test_split(data,label,test_size=0.33, random_state=42)

X_train = x_train
Y_train = y_train

X_test = x_test
Y_test = y_test 

print(X_train[0])
```

```python
max_document_length = max([len(x.split(" ")) for x in X_train])
vocab = learn.preprocessing.VocabularyProcessor(max_document_length)

max_document_length_label = max([len(x.split(" ")) for x in Y_train])
vocab_label = learn.preprocessing.VocabularyProcessor(max_document_length_label)

x = np.array(list(vocab.fit_transform(X_train)))
x1 = np.array(list(vocab.fit_transform(X_test)))


y = LabelEncoder.fit_transform(Y_train)
y1 = LabelEncoder.fit_transform(Y_test)
```

```python
max_features = 100
maxlen = 30
batch_size = 1000
embedding_dims = 250
filters = 64
kernel_size = 3
epoch = 20
```

```python
x = sequence.pad_sequences(x, maxlen= maxlen, padding = 'post')
x1 = sequence.pad_sequences(x1, maxlen= maxlen, padding = 'post')

trainy_enc = to_categorical(y)
testy_enc = to_categorical(y1)

print(x.shape)
print(x1.shape)
```

```python
from keras.layers import Activation

x = x.reshape((334686, 30, 1))
x1 = x1.reshape((164846, 30,1))


model = Sequential()
model.add(LSTM(16, return_sequences=True, input_shape=(30, 1)))
model.add(LSTM(16, return_sequences=True))
model.add(LSTM(16))
model.add(Dense(6, activation = 'softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
start_time = time.time()
hist = model.fit(x, trainy_enc, batch_size=batch_size, epochs = epoch, shuffle = True, validation_data=(x1, testy_enc))
elapsed_time = time.time() - start_time
print("processing time: ", elapsed_time, "Seconds")
```

```python
import matplotlib.pyplot as plt

plt.plot(hist.history['acc'])
plt.plot(hist.history['val_acc'])
plt.title("Model accuracy")
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train','test'], loc = 'upper left')
plt.show()

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title("Model loss")
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['train','test'], loc = 'upper left')
plt.show()
```

```python
import seaborn as sn
import pandas as pd 
from numpy  import array

_, train_acc = model.evaluate(x, trainy_enc, verbose=0)
_, test_acc = model.evaluate(x1, testy_enc, verbose=0)

print('Train: %.3f, Test: %.3f' % (train_acc, test_acc)) 
```
